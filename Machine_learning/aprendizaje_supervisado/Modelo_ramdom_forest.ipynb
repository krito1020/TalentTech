```python
try:
    # Crear el dataframe:
    df = pd.read_excel(file_path)

    #1. Realizar el proceso de limpieza del dataframe
    if 'Unnamed: 9' in df.columns:
        df = df.drop(columns = ['Unnamed: 9'])
        print ('Se ha eliminado la columna "Unnamed: 9"')

    #2. Eliminar filas con valores nulos:
    initial_rows = df.shape[0]
    df = df.dropna() #cuando es un metodo debe ir despues de la palabra parentesis
    row_dropped = initial_rows - df.shape[0]
    print(f'Hemos eliminado {row_dropped} filas con valores nulos')

    #3. Catear adecuadamente el tipo de dato: en este caso fecha
    df['Fecha de Venta'] = pd.to_datetime(df['Fecha De Venta'])
    print(f'Columna "Fecha de Venta" transformada a formato de fecha datatime')
    
    #4. Extraer las caracteristicaas de la fecha
    df['Año de Venta'] = df['Fecha de Venta'].dt.year
    df['Mes de Venta'] = df['Fecha de Venta'].dt.month
    df['Día de semana de Venta'] = df['Fecha de Venta'].dt.dayofweek # Lunes=0.... Domingo=6
    print(f'Caracteristicas de fecha (Año, Mes, Día de la semana), extraidas')

    #5. Impresiones:
    print("\nDataframe después de limpieza, primeras 5 filas:")
    print(df.head().to_markdown(index=False, numalign='left', stralign='left')) 
    print('\nInformación del dataframe después de la limpieza')
    print(df.info())
    print('\nValores Nulos despúes de la limpieza')
    print(df.isnull().sum().to_markdown(index=False, stralign='left'))

    #6. Guardar el dataframe en un archivo más eficiente para que no pese tanto
    df.to_pickle('./datasets/DB_Ventas_Tech_cleaned.pkl')
    print('Dataframe guardado en dataset "DB_Ventas_Tech_cleaned.pkl"')

except Exception as e:
    print(f"Error durante la limpieza y preparación {e}")  
```

```python
    # 1. Empezamos con la ingeneria del modelo
try:
        df = pd.read_pickle(file_path)
        print(f'Cargado el dataframe desde {file_path}, exitosamente')
        display(df)
    
        # a. Empezamos con la ingenería de caracteristicas
        df['Margen de Ganancia'] = df['Precio Unidad'] - df['Costo Unidad'] #Margen de ganacia por unidad
        print (f'Caracteristica \'Margen de Ganancia\' creada')
    
        #b. Ingenería de caracteristicas: ingreso total de venta
        df['Ingreso Total Venta'] = df['Precio Unidad'] * df['Cantidad Vendida']
        print(f'Caraceristica \'Ingreso Total Venta\' creada')
           
        #c. Ingenería de caracteristicas: Ganancia total de venta
        df['Ganancia Total Venta'] = df['Margen de Ganancia'] * df['Cantidad Vendida']
        print(f"Caracteristica \'Ganancia Total Venta\' creada")
    
        # d. Selección de la variable objetivo
        target_variable = 'Ingreso Total Venta'
    
    # Excluimos las columnas que no se usan y las que usamos para calcular el target_variable, e incluso  esta misma para evitar el sobre costeo de data (data leak)
     
    # Caracteristicas a descartar:
        feacture_to_exclude = ['Fecha de venta', 'Nombre Cliente', target_variable, 'Precio Unidad', 'Cantidad Vendida']
        feacture = [col for col in df.columns if col not in feacture_to_exclude]
       
    # Identificar caracteristicas categoricas y númericas:
        categorical_feacture = df[feacture].select_dtypes(include = ['object']).columns
        numerical_feacture = df[feacture].select_dtypes(include = ['int64', 'float64', 'int32']).columns
       
    #imprime el resultado:
        print(f'Variable objetivo: {target_variable}')
        print(f'Caracteristicas Categoricas: {list(categorical_feacture)}')
        print(f'Caracteristicas Númericas: {list(numerical_feacture)}')
    
        # e. Codificar variables categoricas (One-Hot-Encoding) y pasar númericas:
        preprocessor = ColumnTransformer(
            transformers = [
                ('num', 'passthrough', numerical_feacture),
                ('cat', OneHotEncoder(handle_unknown = "ignore"),categorical_feacture)
            ]
        )
    
        # f. Aplicar transformaciones:
        x = df[feacture]
        y = [target_variable]
    
        x_processed = preprocessor.fit_transform(x)
    
    # g. Convertir en un arreglo denso si es una matriz dispersa:
        if scipy.sparse.issparse(x_processed):
            x_processed = x_processed.toarray()
    
    # h. Obtenemos los nombres de las nuevas columnas despues del preprocesamiento:
        all_feacture_names = preprocessor.get_feacture_names_out()
    
        x_processed_dataframe = pd.DataFrame(x_processed, columns = all_feacture_names, index = df.index)
        #impresiones
        print (f'Dataframe después de la ingeniera de caracteristicas y codificación:')
        display(x_processed_Dataframe.head())
        print(f'Dimensiones del dataset despues de la ingeniera de caracteristicas y codificación:')
        print(x_processed_dataframe.shape)
    
    # i. Guardar el modelo generado:
        joblib.dump(preprocessor, './Modelo/pre_rf_model.pkl')
    # j. Guardar
        x_processed_dataframe.to_pickle('./datasets/x_processed_dataframe_df.pkl')
        
except Exception as e:
        print(f'Error presentado al guardar el modelo {e}')
```
