```python
try:
    # crear el dataframe:
    df = pd.read_excel(file_path)

    # 1. Realizar el proceso de limpieza del dataframe:
    if 'Unnamed: 9' in df.columns:
        df = df.drop(columns = ['Unnamed: 9'])
        print('Se ha eliminidado la columna "Unnamed: 9"')

    # 2. Eliminar Filas con valores nulos:
    initial_rows = df.shape[0]
    df = df.dropna()
    row_dropped = initial_rows - df.shape[0]
    print(f'Hemos Eliminado {row_dropped} filas con valores nulos')

    # 3. Casteamos adecuadamente el tipo de dato:
    df['Fecha de Venta'] = pd.to_datetime(df['Fecha De Venta'])
    print(f'Columna "Fecha de Venta" transformada a formato de fecha datetime')

    # 4. Extraer las carateristicas de la fecha
    df['Año de Venta'] = df['Fecha de Venta'].dt.year
    df['Mes de Venta'] = df['Fecha de Venta'].dt.month
    df['Día de Semana de Venta'] =  df['Fecha de Venta'].dt.dayofweek  # lunes = 0 .... Domingo = 6
    print(f'Caracteristicas de fecha (Año, Mes, Día de la semana), extraidas')

    # Impresiones:
    print("\nDataframe después de limpieza, primeras 5 filas:")
    print(df.head().to_markdown(index=False, numalign='left', stralign='left'))
    print('\nInformación del dataframe después de la limpieza')
    print(df.info())
    print('\nValores Nulos por columna después de la limpieza')
    print(df.isnull().sum().to_markdown(index=False, stralign='left'))

    # Guardamos el dataset resultante:
    df.to_pickle('./datasets/BD_Ventas_Tech_cleaned.pkl')
    print('Dataframe guardado en dataset "BD_Ventas_Tech_cleaned.pkl"')

except Exception as e:
    print(f"Error durante la limpieza y preparación {e}")
```

## Paso 3: Ingenieria de Modelo:

```python
from sklearn.preprocessing import OneHotEncoder          # 
from sklearn.compose import ColumnTransformer
import joblib
import scipy.sparse

# librerias para el entrenammiento de la ml:
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
file_path = './datasets/BD_Ventas_Tech_cleaned.pkl'

try:
    df = pd.read_pickle(file_path)
    print(f'Cargado el dataframe desde {file_path}, exitosamente')

    # 1. Empezamos con la ingeniería de caracteristicas:
    df['Margen Ganancia'] = df['Precio Unidad'] - df['Costo Unidad']
    print(f"Caracteristica \'Margen de Ganancia\' creada")

    # 2. Ingeniería de caracteristicas: ingreso total de venta
    df['Ingreso Total Venta'] = df['Precio Unidad'] * df['Cantidad Vendida']
    print(f"Caracteristica \'Ingreso Total Venta\' creada")
    
    # 3. Ingeniería de caracteristicas: Ganancia total de venta
    df['Ganancia Total Venta'] = df['Margen Ganancia'] * df['Cantidad Vendida']
    print(f"Caracteristica \'Ganancia Total Venta\' creada")

    # 4. Selección de la variable objetivo:
    target_variable = 'Ingreso Total Venta'

    # Excluimos las columnas que no se usan y las que usamos para calcular el target_variable, e incluos esta misma para evitar el data leake
    # Caracteristicas a descartar:
    features_to_exclude = ['Fecha de Venta', 'Nombre Cliente', target_variable, 'Precio Unidad', 'Cantidad Vendida']
    
    features = [col for col in df.columns if col not in features_to_exclude]
    print(f'Caracteristicas para ingeniería: {features}')

    # Identificar caracteristicas categoricas y numéricas:
    categorical_features = df[features].select_dtypes(include = ['object']).columns
    numerical_features = df[features].select_dtypes(include = ['int64','float64','int32']).columns

    # Imprimamos el resultado:
    print(f"Variable Objetivo: {target_variable}")
    print(f"Caracteristicas Categoricas: {list(categorical_features)}")
    print(f"Caracteristicas Numéricas: {list(numerical_features)}")

    # 5. Codificar variables categorias (One-Hot-Encoding) y pasar númericas:
    preprocessor = ColumnTransformer(
        transformers = [
            ("num", "passthrough", numerical_features),
            ("cat", OneHotEncoder(handle_unknown="ignore"),categorical_features)
        ]
    )

    # Aplicar Transformaciones:
    x = df[features]
    y = df[target_variable]

    x_processed = preprocessor.fit_transform(x)

    # Convertir en un arreglo denso si es una matriz dispersa:
    if scipy.sparse.issparse(x_processed):
        x_processed = x_processed.toarray()

    # Obtenemos los nombres de las nuevas columnas despues del preprocesamiento:
    all_features_names = preprocessor.get_feature_names_out()

    x_processed_dataframe = pd.DataFrame(x_processed, columns = all_features_names, index = df.index)

    print(f"Dataframe después de la ing de caracteristicas y codificación:")
    display(x_processed_dataframe.head())
    print(f"Dimensiones del dataset despues de todo esto:")
    print(x_processed_dataframe.shape)

    # Guardar el modelo generado:
    joblib.dump(preprocessor, "./Modelo/pre_rf_model.pkl")
    # Guardar el nuevo dataset
    x_processed_dataframe.to_pickle('./datasets/x_processed_dataframe_df.pkl')
    y.to_pickle('./datasets/y_target.pkl')

except Exception as e:
    print(f'error presentado al guardar el modelo {e}')
```

# Paso 4: Entrenamiento y evaluaciòn del modelo:

```python
# Cargar los dataframes procesados:
x_processed_dataframe = pd.read_pickle('./datasets/x_processed_dataframe_df.pkl')
y_target = pd.read_pickle('./datasets/y_target.pkl')

# division para entrenamiento y evaluación
x_train, x_test, y_train, y_test = train_test_split(x_processed_dataframe, y_target, test_size=0.2, random_state=42)

print(f"Dimensiones del conjunto de pruebas: x_test = {x_test.shape}, y_test:  {y_test.shape}")
print(f"Dimensiones del conjunto de entrenamiento: x_test = {x_train.shape}, y_test:  {y_train.shape}")
```

# Iniciamos el entrenamiento

```python
print(f"Iniciando el entrenamiento")
model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(x_train, y_train)
print(f"Entrenamiento concluido con éxito")
```

# Evaluando el modelo

```python
# Predicciones:
y_pred = model.predict(x_test)

# Evaluación de lo predicho:
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

## Imprimir los resultados:

```python
print(f'\nResultados de la evaluación del modelo')
print(f'Error cuadratico medio (MSE): {mse:.2f}')
print(f'Coeficiente de determinación (R2): {r2:.2f}')
```

# Guardamos el modelo

```python
joblib.dump(model, './Modelo/model_train_rf.pkl')
print('Hemos Guardado el Modelo Entrenado')
```

##Optimización del video

```python
# Cargar los dataframes procesados:
x_processed_dataframe = pd.read_pickle('./datasets/x_processed_dataframe_df.pkl')
y_target = pd.read_pickle('./datasets/y_target.pkl')

# division para entrenamiento y evaluación
x_train, x_test, y_train, y_test = train_test_split(x_processed_dataframe, y_target, test_size=0.2, random_state=42)

print(f"Dimensiones del conjunto de pruebas: x_test = {x_test.shape}, y_test:  {y_test.shape}")
print(f"Dimensiones del conjunto de entrenamiento: x_test = {x_train.shape}, y_test:  {y_train.shape}")
```

```python
model = RandomForestRegressor(random_state=42, n_jobs=-1)

param_grid = {
    'n_estimators': [50, 100, 150], #Arboles de decisión a entrenar
    'max_depth': [10, 20, None] # None significa que no hay límite de profundidad  -Profundidad del bosque
}
```

```python
print('iniciando la búsqueda de hiperparámetros con GridSearchCV')
grid_search = GridSearchCV(estimator = model, param_grid= param_grid, cv=3, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error') #n_jobs=-1 para usar todos los núcleos disponibles - verbose=2 para ver el progreso

grid_search.fit(x_train, y_train)

print('Busqueda de hiperparámetros completada')
```

```python
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print(f'\nMejor conjunto de hiperparámetros encontrados: {best_params}')
```

##..... Evaluación y creación de mejor modelo .....

```python
y_pred_optimized = best_model.predict(x_test)

mse_optimized = mean_squared_error(y_test, y_pred_optimized) # Error cuadrático medio (MSE)
r2_optimized = r2_score(y_test, y_pred_optimized) #Coeficiente de determinación (R2)

#Impresiones del resultado:
print(f'\nResultados de la evaluación del modelo')
print(f'Error cuadrático medio (MSE): {mse_optimized:.2f}')
print(f'Coeficiente de determinación (R2): {r2_optimized:.2f}')
```

##..... Guardamos el modelo .....

```python
joblib.dump(best_model, './Modelo/best_model_rf.pkl')
print('Hemos Guardado el Modelo Optimizado Entrenado' )
```
