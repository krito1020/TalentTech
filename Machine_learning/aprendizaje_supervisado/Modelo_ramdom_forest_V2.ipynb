## 2do Paso: Recolectar Datos:

```python
import pandas as pd
```

```python
v_2018 = './datasets/ventas_2018.xlsx' # Cargamos los dataframes de ventas de los años 2018 a 2022
v_2019 = './datasets/ventas_2019.xlsx'
v_2020 = './datasets/ventas_2020.xlsx'
v_2021 = './datasets/ventas_2021.xlsx'
v_2022 = './datasets/ventas_2022.xlsx'

df_1 = pd.read_excel(v_2018, header=None) 
df_2 = pd.read_excel(v_2019, header=None)
df_3 = pd.read_excel(v_2020, header=None)
df_4 = pd.read_excel(v_2021, header=None)
df_5 = pd.read_excel(v_2022, header=None)


df_ventas_previa = pd.concat([df_1, df_2, df_3, df_4, df_5], axis=0, ignore_index=True) # Concatenamos los dataframes de ventas de los años 2018 a 2022
df_ventas_final = pd.concat([df_1, df_ventas_previa], axis=0, ignore_index= True) 
# axis=0 para concatenar por filas, axis=1 para concatenar por columnas
# ignore_index=True para ignorar los índices originales y crear un nuevo índice, inner=outer para mantener solo las columnas que están en ambos dataframes
display(df_ventas_final)
```
