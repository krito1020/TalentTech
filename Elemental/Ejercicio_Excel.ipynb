```python
import pandas as pd #Para importar pandas. Lo puedo renombrar con el AS
```

```python
# *1.* crear vairbale, de la ruta del archivo
path_xlsx = ('./BD/BD_Ventas_Tech.xlsx')
```

```python
# 2. montar el dataframe
df_xlsx = pd.read_excel(path_xlsx)
```

```python
display(df_xlsx) #muestra la información que se leyo por el dataframe
```

*1.* Validación de columna (nombre y cantidad de columnas)

```python
display(df_xlsx.columns) #muestra las columnas que tiene el dataframe
```

*2.* Ejecutar shape e info

```python
print("\n" + "="*50 + "\n")
print(f"El dataframe df_xlsx, tiene {df_xlsx.shape[0]} filas, y {df_xlsx.shape[1]} columnas") #muestra la cantidad de filas y columnas

print("\n" + "="*50 + "\n")
print(f"Clasificación de las columnas de dataset")
df_xlsx.info() #muestra la información del dataframe, como el tipo de dato de cada columna
```

*3.* Validar tipo de columnas

```python
display(df_xlsx.dtypes)
```

4. La estadistica descriptiva del data set

```python
display(df_xlsx.describe()) #muestra un resumen estadístico
```

5. Hace analisis estadistico pero incluyendo los objetos, es decir estadistica cualitativa

```python
display(df_xlsx.describe(include= 'object')) #muestra un resumen estadístico
```
